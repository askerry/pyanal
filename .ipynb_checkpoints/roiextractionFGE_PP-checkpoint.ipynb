```python
pdfs=[pd.read_csv(f) for f in pscfiles]
pdf=pd.concat(pdfs)
bdfs=[pd.read_csv(f) for f in betafiles]
bdf=pd.concat(bdfs)
TRcols=[col for col in pdf.columns if 'TR' in col]
```

```python
#dropping bad trials
try:
    for subj in droprunconds.keys():
        for run in droprunconds[subj].keys():
            for cond in droprunconds[subj][run]:
                pdf.loc[(pdf['subject']==subj) & (pdf['runs']==run) & (pdf['condition']==cond)][TRcols]=np.nan
                bdf.loc[(bdf['subject']==subj) & (bdf['condition']==cond)]['Run %s' %run]=np.nan
except:
    pass
```

```python
runavgs=pdf.groupby(['roi','condition', 'subject']).mean().reset_index()
runavgs['runs']='all'
runavgs['roi']=[roi[:roi.index('_')] for roi in runavgs['roi'].values]
```

```python
#manual averaging
def averagedfs(hstacked, groupby=['roi', 'condition'], keepcols=None):
    if not keepcols:
        keepcols=hstacked.columns
    condmeans=runavgs.groupby(groupby).mean()
    condstds=runavgs.groupby(groupby).std()
    condcounts=runavgs.groupby(groupby).count()
    condsems=condstds[keepcols].values/(np.sqrt(condcounts[keepcols])).values
    conderror=condstds.copy()
    conderror.ix[:,keepcols]=condsems
    return condmeans, conderror
cmeans, cerror= averagedfs(runavgs, groupby=['roi', 'condition'], keepcols=['artifacts removed']+TRcols)
```

```python
def verticalizetimeseries(hstacked, timecols, unit, othercols=[]):
    othervals=[hstacked.reset_index()[col].values for col in othercols]
    allcols=othercols+[unit, 'value', 'time', 'timename']
    vstacked=pd.DataFrame(columns=allcols, index=range(len(hstacked)*len(TRcols)))
    for coln,col in enumerate(timecols):
        crange=np.arange(len(hstacked))+len(hstacked)*coln
        value=hstacked[col].values
        units=hstacked[unit].values
        timename=[float(col[:col.index(' ')]) for el in range(len(hstacked))]
        time=[coln for el in range(len(hstacked))]
        data=np.array(othervals + [units, value, time, timename]).T
        vstacked.ix[crange,:]=data
    for c in othercols:
        vstacked=vstacked[vstacked[c].notnull()]
    return vstacked
tseries=verticalizetimeseries(runavgs, TRcols, 'subject', othercols=['roi','condition'])
```

```python
import seaborn as sns
roilist=tseries.roi.unique()
roiseries=tseries[tseries['roi']==roilist[0]]
ax=sns.tsplot(roiseries, value='value', condition='condition', unit='subject', time='time', color=colors)
```
